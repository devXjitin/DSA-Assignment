# Sorting Algorithms - Concepts and Working

## 1. Merge Sort

### Concept
- Merge Sort is a divide-and-conquer algorithm that divides the array into smaller subarrays, sorts them, and then merges them back together.
- It recursively splits the array in half until each subarray contains a single element.
- Then it repeatedly merges the subarrays to produce new sorted subarrays until only one sorted array remains.

### How It Works
1. **Divide**: Split the array into two halves
2. **Conquer**: Recursively sort both halves
3. **Combine**: Merge the two sorted halves back together

### Example
```
Initial Array: [38, 27, 43, 3, 9, 82, 10]

Step 1 (Divide):
[38, 27, 43, 3] | [9, 82, 10]
[38, 27] [43, 3] | [9, 82] [10]
[38] [27] [43] [3] | [9] [82] [10]

Step 2 (Merge):
[27, 38] [3, 43] | [9, 82] [10]
[3, 27, 38, 43] | [9, 10, 82]
[3, 9, 10, 27, 38, 43, 82]
```

### Characteristics
- **Stable**: Maintains relative order of equal elements
- **Not In-place**: Requires additional memory for temporary arrays
- **Consistent Performance**: Always O(n log n) regardless of input

## 2. Quick Sort

### Concept
- Quick Sort is a divide-and-conquer algorithm that selects a 'pivot' element and partitions the array around it.
- Elements smaller than the pivot go to the left, and elements larger go to the right.
- This process is recursively applied to the sub-arrays.

### How It Works
1. **Choose a pivot**: Select an element from the array (often the middle or last element)
2. **Partition**: Rearrange the array so elements less than the pivot come before it, and elements greater come after
3. **Recursively sort**: Apply the same process to the left and right subarrays

### Example
```
Initial Array: [10, 7, 8, 9, 1, 5]
Pivot = 5

Step 1 (Partition around 5):
[1] 5 [10, 7, 8, 9]

Step 2 (Partition left and right):
[1] 5 [7, 8, 9] 10
[1] 5 [7] 8 [9] 10

Final: [1, 5, 7, 8, 9, 10]
```

### Characteristics
- **Not Stable**: May change relative order of equal elements
- **In-place**: Uses minimal extra memory
- **Variable Performance**: Average case O(n log n), worst case O(n²) if pivot selection is poor

## 3. Selection Sort

### Concept
- Selection Sort divides the array into a sorted and unsorted region.
- It repeatedly finds the minimum element from the unsorted region and places it at the beginning of the sorted region.
- Simple but inefficient for large datasets.

### How It Works
1. **Find minimum**: Find the smallest element in the unsorted portion
2. **Swap**: Swap it with the first element of the unsorted portion
3. **Move boundary**: Expand the sorted portion by one element
4. **Repeat**: Continue until the entire array is sorted

### Example
```
Initial Array: [64, 25, 12, 22, 11]

Pass 1: [11, 25, 12, 22, 64]  (11 is minimum, swap with 64)
Pass 2: [11, 12, 25, 22, 64]  (12 is minimum, swap with 25)
Pass 3: [11, 12, 22, 25, 64]  (22 is minimum, swap with 25)
Pass 4: [11, 12, 22, 25, 64]  (25 is minimum, already in place)

Final: [11, 12, 22, 25, 64]
```

### Characteristics
- **Not Stable**: May change relative order of equal elements
- **In-place**: Requires only constant extra memory
- **Minimum Swaps**: Makes only O(n) swaps, useful for write-limited memory

## 4. Insertion Sort

### Concept
- Insertion Sort builds the sorted array one element at a time.
- It picks each element and inserts it into its correct position in the already sorted portion.
- Works similarly to how people sort playing cards in their hands.

### How It Works
1. **Start**: Assume the first element is sorted
2. **Pick next**: Take the next element from the unsorted portion
3. **Compare**: Compare it with elements in the sorted portion (moving backwards)
4. **Shift**: Shift larger elements one position to the right
5. **Insert**: Place the element in its correct position
6. **Repeat**: Continue until all elements are sorted

### Example
```
Initial Array: [12, 11, 13, 5, 6]

Pass 1: [12, 11, 13, 5, 6]  (12 is sorted)
Pass 2: [11, 12, 13, 5, 6]  (Insert 11 before 12)
Pass 3: [11, 12, 13, 5, 6]  (13 stays in place)
Pass 4: [5, 11, 12, 13, 6]  (Insert 5 at beginning)
Pass 5: [5, 6, 11, 12, 13]  (Insert 6 after 5)

Final: [5, 6, 11, 12, 13]
```

### Characteristics
- **Stable**: Maintains relative order of equal elements
- **In-place**: Requires only constant extra memory
- **Adaptive**: Efficient for nearly sorted data (O(n) in best case)

## 5. Bubble Sort

### Concept
- Bubble Sort repeatedly steps through the array, compares adjacent elements, and swaps them if they're in the wrong order.
- The largest elements "bubble up" to their correct position at the end with each pass.
- Simple to understand but inefficient for large datasets.

### How It Works
1. **Compare**: Compare adjacent elements in the array
2. **Swap**: If they're in wrong order, swap them
3. **Move**: Move to the next pair
4. **Repeat pass**: After each complete pass, the largest unsorted element is in its final position
5. **Continue**: Repeat until no swaps are needed

### Example
```
Initial Array: [5, 1, 4, 2, 8]

Pass 1:
[1, 5, 4, 2, 8] (swap 5 and 1)
[1, 4, 5, 2, 8] (swap 5 and 4)
[1, 4, 2, 5, 8] (swap 5 and 2)
[1, 4, 2, 5, 8] (8 is in place)

Pass 2:
[1, 4, 2, 5, 8] (no swap)
[1, 2, 4, 5, 8] (swap 4 and 2)
[1, 2, 4, 5, 8] (5 is in place)

Pass 3:
[1, 2, 4, 5, 8] (no swaps needed)

Final: [1, 2, 4, 5, 8]
```

### Characteristics
- **Stable**: Maintains relative order of equal elements
- **In-place**: Requires only constant extra memory
- **Adaptive**: Can detect if array is already sorted (with optimization)

## Key Differences Summary

| Algorithm | Time Complexity (Avg) | Space | Stable | Best Use Case |
|-----------|----------------------|-------|--------|---------------|
| Merge Sort | O(n log n) | O(n) | Yes | Large datasets, guaranteed performance |
| Quick Sort | O(n log n) | O(log n) | No | General purpose, fast average case |
| Selection Sort | O(n²) | O(1) | No | Small datasets, minimal memory |
| Insertion Sort | O(n²) | O(1) | Yes | Small or nearly sorted data |
| Bubble Sort | O(n²) | O(1) | Yes | Educational, very small datasets |

**Terms Explained:**
- **Stable**: A sorting algorithm is stable if two elements with equal values appear in the same order in the sorted output as they appear in the input.
- **In-place**: An algorithm is in-place if it uses only a constant amount of extra space.
- **Adaptive**: An algorithm is adaptive if it takes advantage of existing order in its input.

# Performance Graph of Sorting Algorithms

````python
import random
import time
import matplotlib.pyplot as plt


def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1


def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)


def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]


def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key


def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]


# Function to measure execution time
def measure_time(size):
    arr = [random.randint(1, 10000) for _ in range(size)]
    start_time = time.time()
    bubble_sort(arr)
    end_time = time.time()
    return end_time - start_time


sizes = [1000, 3000, 5000, 7000, 10000]
times = {
    "Merge Sort": [],
    "Quick Sort": [],
    "Selection Sort": [],
    "Insertion Sort": [],
    "Bubble Sort": [],
}

for size in sizes:
    arr = [random.randint(0, 100000) for _ in range(size)]

    # Merge Sort
    temp = arr[:]
    start = time.time()
    merge_sort(temp)
    times["Merge Sort"].append(time.time() - start)

    # Quick Sort
    temp = arr[:]
    start = time.time()
    quick_sort(temp)
    times["Quick Sort"].append(time.time() - start)

    # Selection Sort
    temp = arr[:]
    start = time.time()
    selection_sort(temp)
    times["Selection Sort"].append(time.time() - start)

    # Insertion Sort
    temp = arr[:]
    start = time.time()
    insertion_sort(temp)
    times["Insertion Sort"].append(time.time() - start)

    print(f"Completed for size {size}")

    # Bubble Sort
    temp = arr[:]
    start = time.time()
    bubble_sort(temp)
    times["Bubble Sort"].append(time.time() - start)

    print(f"Completed for size {size}")


for algo, t in times.items():
    plt.plot(sizes, t, marker="o", linestyle="--", label=algo)

plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.title("Sorting Algorithm Comparison")
plt.legend()
plt.grid(True)
plt.show()
````

# Which Algorithm is Better?

## For Large Datasets
- **Quick Sort** is usually the fastest in practice with average O(n log n) and low memory usage.
- **Merge Sort** is safer if you need guaranteed O(n log n) time or a stable sort.

## For Small or Nearly Sorted Datasets
- **Insertion Sort** is often fastest because of its simple inner loop and O(n) best case.
- **Avoid Bubble Sort** - it is almost never the best choice, too slow for practical use.

## Overall Best Choice
- **General Purpose**: Quick Sort (fastest on average)
- **When stability or worst-case guarantees matter**: Merge Sort
- **For small or nearly sorted arrays**: Insertion Sort
- **Educational purposes only**: Bubble Sort
- **When dataset is tiny and extra memory must be zero**: Selection Sort

# Real-World Applications

## 1. Quick Sort
- **Use Cases**: Search engines, e-commerce platforms, sorting millions of product prices or search results
- **Database Systems**: Many SQL engines use a variant of Quick Sort for in-memory sorting
- **Why**: Average O(n log n) time complexity, low extra memory usage, very fast for large datasets

## 2. Merge Sort
- **Use Cases**: Email inbox, contact lists requiring stable ordering
- **External Sorting**: Merging large log files or movie libraries stored on disk
- **Why**: Guaranteed O(n log n) even in worst case, and provides stable sorting

## 3. Insertion Sort
- **Use Cases**: Typing suggestions, auto-complete features, small leaderboards
- **Card Games**: Adding a new score into an already-sorted list
- **Why**: O(n) when data is almost sorted, minimal overhead for small datasets

## 4. Selection Sort
- **Use Cases**: Embedded devices, tiny gadgets sorting sensor readings
- **Write-limited Storage**: Only n-1 swaps, reducing wear on storage
- **Why**: Minimal extra memory usage, fewer writes compared to other algorithms

## 5. Bubble Sort
- **Use Cases**: Teaching and debugging, classroom demonstrations
- **Very Small Datasets**: 5-10 items that are nearly sorted except for one element
- **Why**: Trivial to code and visualize, though slow for large inputs

# Summary
In real everyday software:
- **Quick Sort** and **Merge Sort** dominate for general-purpose sorting
- **Insertion Sort** and **Selection Sort** are useful in small or special situations
- **Bubble Sort** is mainly used for educational purposes
