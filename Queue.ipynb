# Queue

- Queue is a linear data structure that follows the First In First Out (FIFO) principle.
- Elements are added at the rear (enqueue) and removed from the front (dequeue).
- Common operations:
  - `enqueue(item)`: Add an item to the rear of the queue.
  - `dequeue()`: Remove and return the item from the front of the queue.
  - `is_empty()`: Check if the queue is empty.
  - `peek()`: Get the front item without removing it.

# Implementation of Queue using Python list

```python
# Implementation of Queue using Python list with operations: enqueue, dequeue, is_full, is_empty, peek, and size.
class Queue:
    def __init__(self, capacity=5):
        self.capacity = capacity
        self.Queue = []

    def is_full(self):
        return len(self.Queue) == self.capacity

    def is_empty(self):
        return len(self.Queue) == 0

    def enqueue(self, item):
        if self.is_full():
            print("Queue is full!")
        else:
            self.Queue.append(item)
            print(f"Enqueued: {item}")

    def dequeue(self):
        if self.is_empty():
            print("Queue is empty!")
            return None
        else:
            item = self.Queue.pop(0)
            print(f"Dequeued: {item}")
            return item

    def peek(self):
        if self.is_empty():
            print("Queue is empty!")
            return None
        else:
            return self.Queue[0]

    def size(self):
        return len(self.Queue)
```

```python
# Example usage
queue = Queue()
print("Is queue empty?", queue.is_empty())
print("Is queue full?", queue.is_full())
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
queue.enqueue(4)
queue.enqueue(5)
queue.enqueue(6)
print("Front element is:", queue.peek())
print("Queue size is:", queue.size())
print("Dequeued element is:", queue.dequeue())
print("Is queue empty?", queue.is_empty())
```

# Time Complexity:
- Enqueue: O(1)
- Dequeue: O(n) (due to shifting elements)
